```python
import pandas as pd
import mplfinance as mpf
```

```python
daily = pd.read_csv('data/SP500_NOV2019_Hist.csv',index_col=0,parse_dates=True)

daily.index.name = 'Date'
daily.shape

daily.head(2)
daily.tail(2)
```

---
Plot horozontal lines.

```python
mpf.plot(daily,hlines=[3080,3121])
```

---
Colour lines and mofidy the style

```python
mpf.plot(daily,hlines=dict(hlines=[3080,3121],colors=['g','r'],linestyle='-.'))
mpf.plot(daily,hlines=dict(hlines=[3080,3121],colors=['g','r'],linestyle='-.'),type='candle')
```

---

Arbitary lines - using a sequence of points to draw lines where a point is (date, price). Arbitary lines can be a sequence of a sequence of points.

```python
df = pd.read_csv('./data/yahoofinance-SPY-20080101-20180101.csv',index_col=0,parse_dates=True)
df = df.loc['2016-05-01':'2016-06-16',:]
df.shape
df.head(2)
df.tail(2)
```

```python
mpf.plot(df)
```

```python
two_points  = [
    ('2016-05-19',203.5),
    ('2016-05-25',209.5)
]

more_points = [
    ('2016-05-02',207),
    ('2016-05-06',204),
    ('2016-05-10',208.5),
    ('2016-05-19',203.5),
    ('2016-05-25',209.5), 
    ('2016-06-08',212),
    ('2016-06-16',207.5)
]

mpf.plot(df,alines=two_points )
mpf.plot(df,alines=more_points)
```
